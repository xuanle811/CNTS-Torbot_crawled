```python
import os
from bs4 import BeautifulSoup

# Đường dẫn thư mục chứa file HTML depth 1
input_dir = 'raw_1/html/'
output_file = 'output/product_links.txt'

```

```python
def extract_product_links(directory, output_folder='outputlink', links_per_file=5):
    product_links = set()
    
    # 1. Tạo thư mục outputlink nếu chưa tồn tại
    if not os.path.exists(output_folder):
        os.makedirs(output_folder)
        print(f"[+] Đã tạo thư mục: {output_folder}")

    if not os.path.exists(directory):
        print(f"[-] Thư mục đầu vào {directory} không tồn tại.")
        return
    
    print(f"[+] Đang quét các file trong: {directory}")
    files = [f for f in os.listdir(directory) if f.endswith(".html")]
    
    for filename in files:
        file_path = os.path.join(directory, filename)
        with open(file_path, 'r', encoding='utf-8') as f:
            soup = BeautifulSoup(f, 'html.parser')
            for a_tag in soup.find_all('a', href=True):
                link = a_tag['href']
                if '/shop/' in link:
                    clean_link = link.split('?')[0]
                    
                    # QUAN TRỌNG: Xóa hết dấu / ở cuối rồi thêm lại ĐÚNG 1 dấu /
                    # Điều này đảm bảo link luôn có dạng .../product-name/
                    clean_link = clean_link.rstrip('/') + '/'
                    if clean_link.endswith('.onion/shop/'):
                        continue
                    product_links.add(clean_link)
            print(f"[*] Đã xử lý: {filename}")

    sorted_links = sorted(list(product_links))
    total_links = len(sorted_links)
    
    if total_links == 0:
        print("[-] Không tìm thấy link sản phẩm nào.")
        return

    print(f"[+] Tìm thấy {total_links} link. Đang chia nhỏ vào thư mục '{output_folder}'...")

    # 2. Lưu file vào thư mục output_folder
    for i in range(0, total_links, links_per_file):
        part_number = (i // links_per_file) + 1
        chunk = sorted_links[i : i + links_per_file]
        
        # Kết hợp đường dẫn thư mục và tên file
        file_path = os.path.join(output_folder, f"product_links_part{part_number}.txt")
        
        with open(file_path, 'w', encoding='utf-8') as f:
            for link in chunk:
                f.write(link + '\n')
        
        print(f"[V] Đã lưu: {file_path}")
```

```python
# run
extract_product_links('raw_depth_1/html/', output_folder='outputlink_all', links_per_file=200)
```
